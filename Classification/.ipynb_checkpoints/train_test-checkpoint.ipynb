```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans

import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8, 6)
pd.options.display.max_rows = 10
```

```python
train = pd.read_csv('TestDataSet.csv', sep=';')
train
```

```python
#test = pd.concat([pd.read_csv('WalkTest.csv'), pd.read_csv('JumpForwardData.csv')])
test = pd.read_csv('WalkTest.csv')
test
```

```python
train.index = train.id
Train = train.drop(['id'], axis=1)
Train = Train.fillna(0)
Train
```

```python
test.index = test.id
Test = test.drop(['id'], axis=1)
Test = Test.fillna(0)
Test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(Train['action'])
Train['label'] = le.transform(Train['action'])
Train = Train.drop(['action'], axis=1)
Train
```

```python
Test['label'] = le.transform(Test['action'])
Test = Test.drop(['action'], axis=1)
Test
```

```python
def to10frames(x):
    result = []
    y = []
    for i in x.video_src.drop_duplicates().tolist():
        for j in range(x[x.video_src == i].values.shape[0]-10):
            result.append(x[x.video_src == i].values[j:j+10,:-2].reshape(360).tolist())
            y.append([x[x.video_src == i].values[j,-1]])
        
    return np.array(result), np.array(y)
```

```python
X_train, y_train = to10frames(Train)
print(X_train)
print(y_train[:10])
```

```python
from pandas.plotting import scatter_matrix

data_numeric = pd.DataFrame(X_train[:,:36], columns=Train.columns[0:-2].tolist())
list_cols = ['Nose_x0', 'RShoulder_x2', 'RHip_x8']
scatter_matrix(data_numeric[list_cols], alpha=0.5, figsize=(10, 10))
plt.show()
```

```python
X_test, y_test = to10frames(Test)
print(X_test)
print(y_test[:10])
```

```python
from pandas.plotting import scatter_matrix

data_numeric = pd.DataFrame(X_test[:,:36], columns=Train.columns[0:-2].tolist())
list_cols = ['Nose_x0', 'RShoulder_x2', 'RHip_x8']
scatter_matrix(data_numeric[list_cols], alpha=0.5, figsize=(10, 10))
plt.show()
```

```python
from sklearn.preprocessing import normalize
X_train = normalize(X_train, 'l2')
X_test = normalize(X_test, 'l2')
print(X_train[:10])
print(X_test[:10])
```

```python
from pandas.plotting import scatter_matrix

data_numeric = pd.DataFrame(X_train[:,:36], columns=Train.columns[0:-2].tolist())
list_cols = ['Nose_x0', 'RShoulder_x2', 'RHip_x8']
scatter_matrix(data_numeric[list_cols], alpha=0.5, figsize=(10, 10))
plt.show()
```

```python
from pandas.plotting import scatter_matrix

data_numeric = pd.DataFrame(X_test[:,:36], columns=Train.columns[0:-2].tolist())
list_cols = ['Nose_x0', 'RShoulder_x2', 'RHip_x8']
scatter_matrix(data_numeric[list_cols], alpha=0.5, figsize=(10, 10))
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, criterion='entropy')
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
predictions = clf.predict(X_test)
```

```python
target_names = le.inverse_transform(list(set(y_train.reshape(y_train.shape[0])))).tolist()
print(classification_report(y_test, predictions, target_names=target_names))

sns.heatmap(confusion_matrix(y_test, predictions), square=True, annot=True, cmap='RdBu', cbar=False,
            xticklabels=target_names,
            yticklabels=target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
```
