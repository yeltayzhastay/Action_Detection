```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans

import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16, 14)
```

```python
df = pd.read_csv('TestDataSet.csv', sep=';')
df
```

```python
df.index = df.id
x = df.drop(['id', 'video_src'], axis=1)
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(x['action'])
x['label'] = le.transform(x['action'])
x = x.drop(['action'], axis=1)
x
```

```python
df.action.describe()
```

```python
def to10frames(x):
    result = []
    y = []
    for i in x.label.drop_duplicates().tolist():
        for j in range(x[x.label == i].values.shape[0]-10):
            result.append(x[x.label == i].values[j:j+10,:-1].reshape(360).tolist())
            y.append([x[x.label == i].values[j,-1]])
        
    return np.array(result), np.array(y)
```

```python
x, y = to10frames(x)
print(x)
print(y)
```

```python
from sklearn import preprocessing
x = preprocessing.normalize(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.632, test_size=0.368, random_state=0)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
pred = clf.predict(X_test)
```

```python
print('Accuracy score: {}'.format(accuracy_score(y_test, pred)))
print('Precision score: {}'.format(precision_score(y_test, pred)))
print('Recall score: {}'.format(recall_score(y_test, pred)))
print('F1 score: {}'.format(f1_score(y_test, pred)))
```

```python
sns.heatmap(confusion_matrix(y_test, pred), square=True, annot=True, cmap='RdBu', cbar=False,
xticklabels=list(set(le.inverse_transform(y))), yticklabels=list(set(le.inverse_transform(y))))
plt.xlabel('true label')
plt.ylabel('predicted label')
```
