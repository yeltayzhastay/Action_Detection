```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DataSetPrimer.csv')
df.index = df.id
df = df.drop(['id', 'action', 'video_src'], axis = 1)
df = df.fillna(0)
df
```

```python
d = df.values
d[:1]
```

```python
R = df.as_matrix()
user_ratings_mean = np.mean(R, axis = 1)
R_demeaned = R - user_ratings_mean.reshape(-1, 1)
R_demeaned
```

```python
user_ratings_mean
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(R_demeaned, k = 1)
```

```python
sigma = np.diag(sigma)
```

```python
user_ratings_mean
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = df.columns)
```

```python
preds_df
```

```python
def vizio(df, pred, _index):
    corr, _ = pearsonr(df.values[_index], pred.values[_index])
    print('Pearsons correlation: %.3f' % corr)
```

```python
for i in range(10):
    vizio(df, preds_df, i)
```
